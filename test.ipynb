```python
from PIL import Image
import torchvision.transforms as T
import torch

# Charger image
img = Image.open("data/texture_ad/wafer/1/test/anomalous/Image_20240709211501692_gray_sx1058_sy958_ps512.png").convert("RGB")

# DÃ©finir les transformations (IMPORTANT)
transform = T.Compose([
    T.Resize((512, 512)),
    T.ToTensor(),  # [0,1], shape: (3, H, W)
    T.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225],
    ),
])

x = transform(img)          # (3, 512, 512)
x = x.unsqueeze(0)          # (1, 3, 512, 512)
```

```python
res=fe(x)

# [
#   feat_0,  # shape: (B, C, H, W)
#   feat_1,  # shape: (B, C, H/2, W/2)
#   feat_2,  # shape: (B, C, H/4, W/4)
# ]
```
